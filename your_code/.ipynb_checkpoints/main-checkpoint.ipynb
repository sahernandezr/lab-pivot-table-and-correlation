#### What's the mean salary by department.

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title

```python
# your answer here
table = pd.pivot_table(employee, index=['Department','Title'], values=['Salary'], aggfunc={'Salary': np.mean})
table
```

#### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table

#### * Hint: Write a custom function*

```python
# your answer here--Your custom function here
def drop_max(employee):
    indexes_to_drop=[]
    max_salaries_by_dept=employee.groupby('Department')['Salary'].idxmax()
    for i in range(0,len(max_salaries_by_dept)):
        indexes_to_drop.append(max_salaries_by_dept[i])
    new_employee = employee.drop(indexes_to_drop, axis=0, inplace=False)
    return new_employee
```

```python
new_employee_no_max = drop_max(employee)
new_employee_no_max
```

```python
pd.pivot_table(new_employee_no_max, values='Salary', index='Department')
```
